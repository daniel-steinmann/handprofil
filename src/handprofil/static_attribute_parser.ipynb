```python
import pandas as pd
import numpy as np
```

```python
measure_labels = pd.read_csv(
    "config/attributes.csv",
    header=0,
    dtype= {
        "id": np.int64,
        "device": str,
        "description": str,
        "unit": str
    }
)
measure_labels
```

```python
info_labels = pd.read_csv(
    "config/meta_attributes.csv",
    header=0,
    dtype= {
        "id": np.int64,
        "description": str,
    }
)
info_labels
```

```python
background = pd.read_csv(
    "config/background.csv",
    header=0,
    dtype= {
        "instrument": str,
        "sex": str,
        "hand": str,
        "id": np.int64,
        "bin_edge": np.int64,
        "value": np.float64
    }
)
background
```

```python
pivot = background.set_index(["instrument","sex"])\
    .pivot_table(index=["instrument","sex","id","bin_edge"], columns="hand")\
    .droplevel(level=0, axis=1)
```

```python
pivot["left"] = pivot["left"].fillna(pivot["right"])
pivot["right"] = pivot["right"].fillna(pivot["left"])
pivot
```
